```python
# Selecting only numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=['int64', 'float64'])

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

data

# It is counting the number of males and females by readint he column 'Gender' from csv file
gender_counts = data['Gender'].value_counts()
explode = (0.05, 0)  #it makes gapes between the chart 
colors = ["#0000FF", "#FF1493"] #Two colors for male and female respectivily
plt.figure(figsize=(8, 8)) #Size of the chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, explode=explode, colors=colors, shadow=True)
plt.title("Gender Distribution") #title
plt.axis('equal')  

plt.show() # For showing graphs
```

```python
data
# It is counting the number of diifrent elements in column "Available in Market" from csv file
gender_counts = data['Available in Market'].value_counts()

explode = (0.08, 0)  # for gaps in chart plots
plt.figure(figsize=(8, 8)) #size of graph
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)
plt.title("Is your preferred food available in market?") #title on chart
plt.axis('equal')  

plt.show() #printing the chart
```

```python
# storing the path od the csv file to a variable df as data frame
df = pd.read_csv("foodDataFile.csv")

fig = px.choropleth(
    df,
    # reading the geo json file for India map from the link
    geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
    #for uniques identification of each Indian state
    featureidkey='properties.ST_NM',
    locations='IndianState',
    #color the state based on the count
    color='StudentCount',
    # red gredient paint
    color_continuous_scale='Reds',
)

fig.update_geos(fitbounds="locations", visible=False)

fig.show()
```
